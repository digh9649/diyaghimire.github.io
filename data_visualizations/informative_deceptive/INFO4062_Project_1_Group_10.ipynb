```python
#importing Oakland

dfOakland = gentrification_df[gentrification_df['city'] == 'Oakland']
dfOakland
```

```python
# Diya
#importing Baltimore
dfBaltimore = gentrification_df[gentrification_df['city'] == 'Baltimore']
gentrified = dfBaltimore[dfBaltimore['gentrified']==True]
drop = 'Census Tract 401, Baltimore city, Maryland'
gentrified = gentrified[gentrified['name'] != drop]
gentrified['total_population_change'] = ((gentrified['total_population_17'] - gentrified['total_population_00']) / gentrified['total_population_00']) * 100
gentrified['name'] = gentrified['name'].str.extract(r'(Census Tract \d+)')
gentrified = gentrified.rename(columns={'total_population_change': 'Change In Total Population'})
gentrified = gentrified.rename(columns={'pct_black_alone_change': 'Change In Black Population'})
```

```python

combined_data = pd.merge(gentrified['Change In Total Population'], gentrified['Change In Black Population'], on=gentrified['name'])
combined_data = combined_data.melt(id_vars='key_0', var_name='dataset', value_name='% Change')
combined_data = combined_data.rename(columns={'key_0':'Census Tracts'})
combined_data['Census Tracts'] = combined_data['Census Tracts'].str.extract(r'( \d+)')

combined_data.loc[16,'Census Tracts']
combined_data = combined_data[combined_data['Census Tracts'] != ' 1308']
```

```python
chart = alt.Chart(combined_data, title='Percent Change in African American and Total Population in Gentrified Tracts (2000-2017)').mark_bar(size=20
    ).encode(
    column = alt.Column('Census Tracts',),
    x =alt.X('dataset', sort = ["Change In Total Population", "Change In Black Population"],  axis=None),
    y =alt.Y('% Change:Q'),
    color= alt.Color('dataset')
).properties(width=55)
```

```python
bar_chart = alt.Chart(combined_data).mark_bar(opacity=0.7).encode(
    x=alt.X('Census Tracts', axis=alt.Axis(title='Census Tracts (Baltimore)',labelFontSize=12, titleFontSize=16)),
    y=alt.Y('% Change', axis=alt.Axis( title='Percent Change', labelFontSize=12, titleFontSize=16)),
    color=alt.Color('dataset', scale=alt.Scale(range=['#1f77b4', '#ff7f0e']), legend=alt.Legend(title="% Change")),
    tooltip=['% Change' ]
).properties(
    width=1000,
    title='Percent Change in African American and Total Population in the Gentrified Tracts of Baltimore (2000-2017)'
)
bar_chart
```

```python
import altair as alt

pop_threshold = 3000
salary_increase_threshold = 10000

popIncomeDF = dfOakland[
    (dfOakland['total_population_00'] > pop_threshold) &
    (dfOakland['total_population_17'] > pop_threshold) &
    (dfOakland['median_income_17'] - dfOakland['median_income_00'] > salary_increase_threshold)
]

popIncomeDF['median_salary_increase'] = popIncomeDF['median_income_17'] - popIncomeDF['median_income_00']

dfOakland_melted = popIncomeDF.melt(
    id_vars=['tract', 'median_salary_increase'],
    value_vars=['pct_white_alone_17', 'pct_black_alone_17', 'pct_asian_alone_17', 'pct_hispanic_or_latino_17'],
    var_name='race',
    value_name='percentage'
)

color_range = ['#4478a6', '#739ebf', '#a2c4d8', '#d0e6f1']

dfOakland_melted['race'] = dfOakland_melted['race'].replace({
    'pct_white_alone_17': 'White %',
    'pct_black_alone_17': 'Black %',
    'pct_asian_alone_17': 'Asian %',
    'pct_hispanic_or_latino_17': 'Hispanic/Latino %'
})

# Construct the chart
chart = alt.Chart(dfOakland_melted).mark_bar().encode(
    x='tract:N',
    y=alt.Y('median_salary_increase:Q', stack='normalize', axis=alt.Axis(format='%'), title='Median Salary Increase ($) 2000-2017'),
    color=alt.Color('race:N', scale=alt.Scale(range=color_range)),
    order=alt.Order('percentage:Q', sort='ascending'),
    tooltip=[
        alt.Tooltip('tract:N', title='Tract'),
        alt.Tooltip('median_salary_increase:Q', title='Median Salary Increase ($)'),
        alt.Tooltip('race:N', title='Group'),
        alt.Tooltip('percentage:Q', title='Population %')
    ]
).properties(
    title='Racial Dispersion and Median Salary Increase (00-17) in Alameda County, CA by Tract'
)

chart
```
